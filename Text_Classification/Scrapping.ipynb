```python
import pandas as pd
import snscrape.modules.twitter as scraper
```

```python
sc=scraper.TwitterUserScraper("WHO")
tweets=[]
for i,tweet in enumerate(sc.get_items()):
    if i>1000:
        break
    tweets.append([tweet.rawContent,"Health"])

df=pd.DataFrame(tweets,columns=['Tweet','Label'])
df.to_csv('WHO.csv')

file=pd.read_csv('WHO.csv')
file
```
